People Dataset

Data Cleaning

```python
# Import Pandas
import pandas as pd
```

```python
# Loading file
people = pd.read_csv('/content/people_dataset.csv')
people
```

```python
# Finding data types
people.info()
```

```python
# Renaming columns
name = {'name':'Name',
        'age':'Age',
        'height':'Height',
        'weight':'Weight',
        'gender':'Gender',
        'country':'Country'
        }

people.rename(columns = name, inplace = True)
people
```

```python
# Changing Gender description
people['Gender'] = people['Gender'].replace(['Male'],'M')
people['Gender'] = people['Gender'].replace(['Female'],'F')

people
```

```python
# Identifying inaccuracies in country column
people['Country'].unique()
```

```python
# Changing countries as England, Northern Ireland, Scotland, Wales
people['Country'] = people['Country'].replace(['Eng land',' England'],'England')
people['Country'] = people['Country'].replace(['Northern Irelnd'],'Northern Ireland')
people['Country'] = people['Country'].replace(['Scot land','Scotland '],'Scotland')
people['Country'] = people['Country'].replace([' Wales'],'Wales')
people
```

```python
# Identifying inaccuracies in age column
people['Age'].unique()
```

```python
# Changing string to number
people['Age'] = people['Age'].replace(['twenty-nine'],29)
people
```

```python
# Dropping index 44
people.drop(index = 44, inplace = True)
people
```

```python
# Converting age column to numerical data
people['Age'] = pd.to_numeric(people['Age'], errors='coerce')

mean_age = people['Age'].mean()
print(format(people['Age'].mean(),'.0f'))
```

```python
# Replacing null values of age with mean age
people['Age'].fillna(mean_age, inplace = True)
people
```

```python
# Average mean height to 0 decimal places
mean_height = people['Height'].mean()
print(format(people['Height'].mean(),'.0f'))
```

```python
# Replacing null values of height with mean height
people['Height'].fillna(mean_height, inplace = True)
people
```

```python
# Calculating total, maximum and minimum age
total_age = people['Age'].sum()
max_age = people['Age'].max()
min_age = people['Age'].min()
print(total_age)
print(max_age)
print(min_age)
```

```python
# Check there are no null values left in any column
people.isnull().sum()
```

Visualisations

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting weight against height per country
x_axis = np.array(people['Weight'])
y_axis = np.array(people['Height'])
plt.scatter(x_axis, y_axis, color = 'purple')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight against Height')
plt.show()
```

```python
countries = ["England", "Scotland", "Wales", "Northern Ireland"]

# Plot one graph per country
for country in countries:
  subset = people[people["Country"] == country]
  plt.figure()
  plt.scatter(subset["Height"], subset["Weight"], color = 'Skyblue')
  plt.xlabel("Height")
  plt.ylabel("Weight")
  plt.title(f"Weight vs Height in {country}")
  plt.show()
```

```python
countries = ["England", "Scotland", "Wales", "Northern Ireland"]

# Plot histogram for each country
for country in countries:
    subset = people[people["Country"] == country]

    plt.figure()
    plt.hist(subset["Weight"].dropna(), bins=8, color = 'darkgreen')
    plt.xlabel("Weight")
    plt.ylabel("Frequency")
    plt.title(f"Weight Distribution in {country}")
    plt.show()
```

```python
countries = ["England", "Scotland", "Wales", "Northern Ireland"]

# Plot histogram for each country
for country in countries:
    subset = people[people["Country"] == country]

    plt.figure()
    plt.hist(subset["Height"].dropna(), bins=8, color = 'green')
    plt.xlabel("Height")
    plt.ylabel("Frequency")
    plt.title(f"Height Distribution in {country}")
    plt.show()
```

```python
countries = ["England", "Scotland", "Wales", "Northern Ireland"]

for country in countries:
    subset = people[people["Country"] == country]

    # Create weight bins
    bins = pd.cut(subset["Weight"], bins=8)

    counts = bins.value_counts().sort_index()

    plt.figure()
    counts.plot(kind="bar")
    plt.xlabel("Weight Range (kg)")
    plt.ylabel("Count")
    plt.title(f"Weight Distribution in {country}")
    plt.xticks(rotation=45)
    plt.show()
```
